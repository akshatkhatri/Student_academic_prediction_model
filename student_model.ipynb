```python
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer,make_column_selector
from sklearn.preprocessing import FunctionTransformer

log_transformer = FunctionTransformer(np.log1p,inverse_func=np.expm1,feature_names_out='one-to-one')

def make_ratio(X):
    return (X[:,0] / (X[:,1] + 1e-8)).reshape(-1,1)

def make_addition(X):
    return (X[:,0] + X[:,1]).reshape(-1,1)

def make_multiplication(X):
    return (X[:,0] * X[:,1]).reshape(-1,1)


def ratio_name(function_transformer,feature_names_in):
    return ["ratio"]

def addition_name(function_transformer,feature_names_in):
    return ["addition"]

def multiplication_name(function_transformer,feature_names_in):
    return ["product"]

ratio_transformer = FunctionTransformer(make_ratio,feature_names_out= ratio_name)
addition_tranformer = FunctionTransformer(make_addition,feature_names_out=addition_name)
multiplication_transformer = FunctionTransformer(make_multiplication,feature_names_out=multiplication_name)

log_pipeline = Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('log',log_transformer),
    ('scale',StandardScaler())
])

ratio_pipeline = Pipeline([
    ("impute",SimpleImputer(strategy='median')),
    ("ratio",ratio_transformer),
    ("scale",StandardScaler())
])

addition_pipeline = Pipeline([
    ("impute",SimpleImputer(strategy='median')),
    ("ratio",addition_tranformer),
    ("scale",StandardScaler())
])

product_pipeline = Pipeline([
    ("impute",SimpleImputer(strategy='median')),
    ("ratio",multiplication_transformer),
    ("scale",StandardScaler())
])

num_pipeline = Pipeline([
    ("impute",SimpleImputer(strategy = 'median')),
    ("scale",StandardScaler())
])

cat_ordi_pipeline = Pipeline([
    ("impute",SimpleImputer(strategy = "most_frequent")),
    ('ordinal_encode',OrdinalEncoder(categories=[
    ['Poor','Fair','Good'],  # diet_quality
    ['High School','Some College','Bachelor','Master','PhD'],  # parental_education_level
    ['Low', 'Medium', 'High'],  # internet_quality
    ['Low', 'Medium', 'High'] ])), # family_income_range
])

cat_hot_pipeline = Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('hot_encode',OneHotEncoder(handle_unknown='ignore'))
])

hot_attri = ['gender','major','part_time_job','extracurricular_participation','dropout_risk','study_environment','access_to_tutoring','learning_style']
ordinal_attri = ['diet_quality','parental_education_level','internet_quality','family_income_range']

preprocessing = ColumnTransformer([
    ('study_time_intensity',product_pipeline,['study_hours_per_day','time_management_score',]),
    ('study_intensity',product_pipeline,['study_hours_per_day','motivation_level']),
    ('chill_hours',addition_pipeline,['netflix_hours','social_media_hours']),
    ('study_to_chill_ratio',ratio_pipeline,['study_hours_per_day','social_media_hours']), # Uncomment after fixing the infinite problem
    ('log',log_pipeline,['study_hours_per_day','sleep_hours','mental_health_rating','stress_level','screen_time']),
    ('ordi',cat_ordi_pipeline,ordinal_attri),
    ('hot',cat_hot_pipeline,hot_attri),
],remainder=num_pipeline)
```
